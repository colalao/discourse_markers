```python
import numpy as np


def print_bootstrap_results(npz_path, language, model):
    with np.load(npz_path, allow_pickle=True) as data:
        
        avg_no_ft_no = data["no_ft_no_bootstrap_avg_scores"]
        avg_ft_no = data["ft_no_bootstrap_avg_scores"]

        # mean and 95% confidence interval
        mean_no_ft_no = np.mean(avg_no_ft_no)
        ci_no_ft_no = np.percentile(avg_no_ft_no, [2.5, 97.5])
        mean_ft_no = np.mean(avg_ft_no)
        ci_ft_no = np.percentile(avg_ft_no, [2.5, 97.5])

        # p-value for the difference
        diffs_no = np.array(avg_ft_no) - np.array(avg_no_ft_no)
        ci_diff_no = np.percentile(diffs_no, [2.5, 97.5])
        p_value_no = 2 * min(np.mean(diffs_no >= 0), np.mean(diffs_no <= 0))
        improvement_no = mean_ft_no - mean_no_ft_no

        avg_no_ft_one = data["no_ft_one_bootstrap_avg_scores"]
        avg_ft_one = data["ft_one_bootstrap_avg_scores"]
        
        mean_no_ft_one = np.mean(avg_no_ft_one)
        ci_no_ft_one = np.percentile(avg_no_ft_one, [2.5, 97.5])
        mean_ft_one = np.mean(avg_ft_one)
        ci_ft_one = np.percentile(avg_ft_one, [2.5, 97.5])
            
        # p-value for the difference
        diffs_one = np.array(avg_ft_one) - np.array(avg_no_ft_one)
        ci_diff_one = np.percentile(diffs_one, [2.5, 97.5])
        p_value_one = 2 * min(np.mean(diffs_one >= 0), np.mean(diffs_one <= 0))
        improvement_one = mean_ft_one - mean_no_ft_one

        print(f"Results for {model} in {language}:") 
        print(f"no_ft_no: mean={mean_no_ft_no:.6f} 95%CI=[{ci_no_ft_no[0]:.4f},{ci_no_ft_no[1]:.6f}]")
        print(f"ft_no   : mean={mean_ft_no:.6f} 95%CI=[{ci_ft_no[0]:.4f},{ci_ft_no[1]:.6f}]")
        print(f"Improvement: {improvement_no:.6f}")
        print(f"Difference 95%CI=[{ci_diff_no[0]:.4f},{ci_diff_no[1]:.4f}], p={p_value_no:.6f}")
        print("--------------------------------------------------")
        print(f"no_ft_one: mean={mean_no_ft_one:.6f} 95%CI=[{ci_no_ft_one[0]:.4f},{ci_no_ft_one[1]:.6f}]")
        print(f"ft_one  : mean={mean_ft_one:.6f} 95%CI=[{ci_ft_one[0]:.4f},{ci_ft_one[1]:.6f}]")
        print(f"Improvement: {improvement_one:.6f}")
        print(f"Difference 95%CI=[{ci_diff_one[0]:.4f},{ci_diff_one[1]:.4f}], p={p_value_one:.6f}")
        

        if model == "LLaMA3" or model == "Qwen3":
            avg_no_ft_full = data["no_ft_full_bootstrap_avg_scores"]
            avg_ft_full = data["ft_full_bootstrap_avg_scores"]

            mean_no_ft_full = np.mean(avg_no_ft_full)
            ci_no_ft_full = np.percentile(avg_no_ft_full, [2.5, 97.5])
            mean_ft_full = np.mean(avg_ft_full)
            ci_ft_full = np.percentile(avg_ft_full, [2.5, 97.5])

            diffs_full = np.array(avg_ft_full) - np.array(avg_no_ft_full)
            ci_diff_full = np.percentile(diffs_full, [2.5, 97.5])
            p_value_full = 2 * min(np.mean(diffs_full >= 0), np.mean(diffs_full <= 0))

            improvement_full = mean_ft_full - mean_no_ft_full

            print("--------------------------------------------------")
            print(f"no_ft_full: mean={mean_no_ft_full:.6f} 95%CI=[{ci_no_ft_full[0]:.4f},{ci_no_ft_full[1]:.6f}]")
            print(f"ft_full  : mean={mean_ft_full:.6f} 95%CI=[{ci_ft_full[0]:.4f},{ci_ft_full[1]:.6f}]")
            print(f"Improvement: {improvement_full:.6f}")
            print(f"Difference 95%CI=[{ci_diff_full[0]:.4f},{ci_diff_full[1]:.4f}], p={p_value_full:.6f}")

        print("===========================================================")
            
            
            
# BERT: Masking
npz_path = "cluster_results/silscore/bert/English_pca100/bootstrap_silscore_layer-1.npz" 
print_bootstrap_results(npz_path, "English", "Bert")

npz_path = "cluster_results/silscore/bert/Japanese_pca100/bootstrap_silscore_layer-1.npz"
print_bootstrap_results(npz_path, "Japanese", "Bert")

# GPT2: NTP and turn-taking
npz_path_1 = "cluster_results/silscore/gpt2/English_pca100/bootstrap_silscore_layer-1.npz"
npz_path_2 = ""
print_bootstrap_results(npz_path_1, "English", "GPT2")

npz_path = "cluster_results/silscore/gpt2/Japanese_pca100/bootstrap_silscore_layer-1.npz"
print_bootstrap_results(npz_path, "Japanese", "GPT2")

# LLaMA3: NTP
npz_path = "cluster_results/silscore/llama3/English_pca100/bootstrap_silscore_layer-1.npz"
print_bootstrap_results(npz_path, "English", "LLaMA3")

npz_path = "cluster_results/silscore/llama3/Japanese_pca100/bootstrap_silscore_layer-1.npz"
print_bootstrap_results(npz_path, "Japanese", "LLaMA3")

# Qwen3: NTP
npz_path = "cluster_results/silscore/qwen3/English_pca100/bootstrap_silscore_layer-1.npz"
print_bootstrap_results(npz_path, "English", "Qwen3")

npz_path = "cluster_results/silscore/qwen3/Japanese_pca100/bootstrap_silscore_layer-1.npz"
print_bootstrap_results(npz_path, "Japanese", "Qwen3")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv("data/Japanese/Japanese_combined_all_files.csv")
file_ids = data['File_ID'].unique()

# 80%, 20%
train_ids, test_ids = train_test_split(file_ids, test_size=0.2, random_state=42)

train_data = data[data['File_ID'].isin(train_ids)]
test_data = data[data['File_ID'].isin(test_ids)]
train_data['File_ID'] = train_data['File_ID'].astype('Int64')
test_data['File_ID'] = test_data['File_ID'].astype('Int64')

train_data.to_csv("data/Japanese/train_data.csv", index=False)
test_data.to_csv("data/Japanese/generate_data.csv", index=False)

print("Train size:", len(train_data))
print("Test size:", len(test_data))

all_file_ids = pd.concat([train_data['File_ID'], test_data['File_ID']]).unique()

train_file_ids = pd.concat([train_data['File_ID']]).unique()
test_file_ids = pd.concat([test_data['File_ID']]).unique()

```

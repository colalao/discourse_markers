```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv("dataset/Japanese/Japanese_combined_all_files.csv")

file_ids = data['File_ID'].unique()

# 75% and 25% 
train_ids, test_ids = train_test_split(file_ids, test_size=0.25, random_state=42)

val_ids, test_ids = train_test_split(test_ids, test_size=0.5, random_state=42)

train_data = data[data['File_ID'].isin(train_ids)]
val_data = data[data['File_ID'].isin(val_ids)]
test_data = data[data['File_ID'].isin(test_ids)]

train_data.to_csv("dataset/Japanese/train_data.csv", index=False)
val_data.to_csv("dataset/Japanese/val_data.csv", index=False)
test_data.to_csv("dataset/Japanese/test_data.csv", index=False)

print("Train size:", len(train_data))
print("Val size:", len(val_data))
print("Test size:", len(test_data))

all_file_ids = pd.concat([train_data['File_ID'], val_data['File_ID'], test_data['File_ID']]).unique()

train_file_ids = pd.concat([train_data['File_ID']]).unique()
val_file_ids = pd.concat([val_data['File_ID']]).unique()
test_file_ids = pd.concat([test_data['File_ID']]).unique()

print("all File_ID :", len(all_file_ids))
print("train File_ID :", len(train_file_ids))
print("val File_ID :", len(val_file_ids))
print("test File_ID :", len(test_file_ids))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 加载数据
data = pd.read_csv("dataset/English/English_combined_all_files.csv")

file_ids = data['File_ID'].unique()
print(len(file_ids))

# 80%, 20%
train_ids, test_ids = train_test_split(file_ids, test_size=0.2, random_state=42)

val_ids, test_ids = train_test_split(test_ids, test_size=0.5, random_state=42)

train_data = data[data['File_ID'].isin(train_ids)]
val_data = data[data['File_ID'].isin(val_ids)]
test_data = data[data['File_ID'].isin(test_ids)]

train_data.to_csv("dataset/English/train_data.csv", index=False)
val_data.to_csv("dataset/English/val_data.csv", index=False)
test_data.to_csv("dataset/English/test_data.csv", index=False)

print("Train size:", len(train_data))
print("Val size:", len(val_data))
print("Test size:", len(test_data))

all_file_ids = pd.concat([train_data['File_ID'], val_data['File_ID'], test_data['File_ID']]).unique()

train_file_ids = pd.concat([train_data['File_ID']]).unique()
val_file_ids = pd.concat([val_data['File_ID']]).unique()
test_file_ids = pd.concat([test_data['File_ID']]).unique()


print("all File_ID :", len(all_file_ids))
print("train File_ID :", len(train_file_ids))
print("val File_ID :", len(val_file_ids))
print("test File_ID :", len(test_file_ids))
```
